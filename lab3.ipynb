```python
#Bai 2.4 Ma trận Markov
import numpy as np

# Ma trận M đầu tiên
M = np.array([[0.8, 0.3], [0.2, 0.7]])
MM = M.dot(M)
print("M^2:")
print(MM)
MM = MM.dot(M)
print("M^3:")
print(MM)
for i in range(100):
    MM = MM.dot(M)
print("M^102:")
print(MM)
# Ma trận M thứ hai
M = np.array([[0.5, 0.5], [0.5, 0.5]])
MM = M.dot(M)
print("M^2 (ma trận đồng đều):")
print(MM)
MM = MM.dot(M)
print("M^3 (ma trận đồng đều):")
print(MM)
for i in range(100):
    MM = MM.dot(M)
print("M^102 (ma trận đồng đều):")
print(MM)
```

```python
#Bai 5
import numpy as np

# Khởi tạo ma trận A (ma trận truyền tin)
A = np.array([
    [0, 1, 0, 1],  # Người 1 gửi cho người 2, 4
    [0, 0, 1, 0],  # Người 2 gửi cho người 3
    [1, 0, 0, 1],  # Người 3 gửi cho người 1, 4
    [1, 1, 0, 0]   # Người 4 gửi cho người 1, 2
])

# Tính A^2
temp = A.dot(A)
print("A^2 (số con đường đi qua 1 trung gian):")
print(temp)

# Tính A^3
temp = temp.dot(A)
print("\nA^3 (số con đường đi qua 2 trung gian):")
print(temp)

# Tính tổng A + A^2 + A^3
sumA = A + A.dot(A) + A.dot(A).dot(A)
print("\nTổng A + A^2 + A^3 (tổng số cách đi qua 0, 1, hoặc 2 trung gian):")
print(sumA)
```

```python
#Bài tập chương
#bai1
import numpy as np
def char_to_num(c):
    if c == ' ':
        return 0
    return ord(c.upper()) - ord('A') + 1
def num_to_char(n):
    if n == 0:
        return ' '
    return chr(n - 1 + ord('A'))
def text_to_numbers(text):
    return [char_to_num(c) for c in text]
def numbers_to_text(nums):
    return ''.join(num_to_char(n) for n in nums)
# (a) Chọn ma trận khả nghịch 3x3
K = np.array([
    [2, -1, 0],
    [1, 1, 1],
    [1, 0, 2]
])
try:
    K_inv = np.linalg.inv(K)
    print("Ma trận K khả nghịch, ma trận nghịch đảo là:")
    print(np.round(K_inv, 3))
except np.linalg.LinAlgError:
    print("Ma trận K không khả nghịch, vui lòng chọn ma trận khác.")
    exit()
# (b) Nhập họ tên hoặc mã số sinh viên
text = input("Nhập họ tên hoặc mã số sinh viên (chỉ chữ cái và khoảng trắng): ").strip()
nums = text_to_numbers(text)
while len(nums) % 3 != 0:
    nums.append(0)
# (c) Mã hóa bằng nhân ma trận
encoded_nums = []
for i in range(0, len(nums), 3):
    block = np.array(nums[i:i+3])
    encoded_block = block.dot(K)
    encoded_nums.extend(encoded_block.astype(int))
print("\nDãy số sau khi mã hóa:")
print(encoded_nums)
# (d) Giải mã bằng ma trận nghịch đảo
decoded_nums = []
for i in range(0, len(encoded_nums), 3):
    block = np.array(encoded_nums[i:i+3])
    decoded_block = block.dot(K_inv)
    decoded_block = np.round(decoded_block).astype(int)
    decoded_nums.extend(decoded_block)
decoded_text = numbers_to_text(decoded_nums)

print("\nChuỗi sau khi giải mã (phục hồi):")
print(decoded_text)

```
